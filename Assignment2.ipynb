# Problem 2

```python
import pandas as pd
pd.set_option("display.max_columns", 100)
url = "https://raw.githubusercontent.com/cs6220/cs6220.spring2019/master/data/AmesHousing.txt"
df1 = pd.read_csv(url, sep="\t")
```

```python
df1.head()
```

# 2.1 Imputation

```python
null_series = df1.isnull().any()
```

```python
count = 0;
null_list = [];
for index, val in null_series.iteritems():
    if val:
        print(index, val)
        count += 1
        null_list.append(index)
print(count, "features have missing values")
```

```python
df_filled = df1.copy()
obj_null_list = []
num_null_list = []
for i in null_list:
    if df_filled[i].dtype == object:
        df_filled[i].fillna("Missing", inplace = True)
        obj_null_list.append(i)
    else:
        num_null_list.append(i)
```

```python
df_filled.head()
```

```python
df_interpolate = df_filled.copy()
```

```python
df_interpolate.interpolate(inplace = True)
```

```python
df_interpolate.head()
```

# 2.2 Standardization

```python
from sklearn import preprocessing
```

```python
df1_transform = df_interpolate.copy()
```

```python
for column in df1_transform.columns:
    if df1_transform[column].dtype == type(object):
        le = preprocessing.LabelEncoder()
        df1_transform[column] = le.fit_transform(df1_transform[column])
```

```python
df1_transform.head()
```

```python
df1_standard = pd.DataFrame(preprocessing.scale(df1_transform),columns=df1_transform.columns)
```

```python
only_num = ['MS SubClass',
 'Lot Frontage',
 'Lot Area',
 'Overall Qual',
 'Overall Cond',
 'Year Built',
 'Year Remod/Add',
 'Mas Vnr Area',
 'BsmtFin SF 1',
 'BsmtFin SF 2',
 'Bsmt Unf SF',
 'Total Bsmt SF',
 '1st Flr SF',
 '2nd Flr SF',
 'Low Qual Fin SF',
 'Gr Liv Area',
 'Bsmt Full Bath',
 'Bsmt Half Bath',
 'Full Bath',
 'Half Bath',
 'Bedroom AbvGr',
 'Kitchen AbvGr',
 'TotRms AbvGrd',
 'Fireplaces',
 'Garage Yr Blt',
 'Garage Cars',
 'Garage Area',
 'Wood Deck SF',
 'Open Porch SF',
 'Enclosed Porch',
 '3Ssn Porch',
 'Screen Porch',
 'Pool Area',
 'Misc Val',
 'Mo Sold',
 'Yr Sold',
 'SalePrice']
```

```python
df1_standard[only_num].head()
```

```python
import matplotlib.pyplot as plt
```

```python
num_list = []
obj_list = []
for col in df1:
    if df1[col].dtype != object:
        num_list.append(col)
        plt.figure()
        df1_transform[[col]].boxplot(grid=False, return_type='axes')
    else:
        obj_list.append(col)
```

```python
for i in num_list:
        plt.figure()
        df1_standard[[i]].boxplot(grid=False, return_type='axes')
```

Overall spreads of boxplots look identical before and after feature standardization. However, mean of each feature varies before standardization; mean of each feature is 0 after standardization which makes it easier to obeserve and identify potential or true outliers.

Sale Price has the outlier furthest from the mean befroe standardization while Misc Val has the outlier furthest from the mean after standardization based on the boxplots generated above.

# 2.3 Feature Selection

```python
from sklearn.feature_selection import mutual_info_regression
from sklearn.feature_selection import SelectKBest
```

```python
num_list_copy = list(num_list)
```

```python
num_list_copy.remove('Order')
```

```python
num_list_copy.remove('PID')
```

```python
num_list_copy.remove('SalePrice')
```

```python
num_list_copy
```

```python
df1_mutal = df1_standard[num_list_copy].copy()
```

```python
df1_mutal.head()
```

```python
X = df1_mutal
```

```python
X.head()
```

```python
y = df1_standard["SalePrice"]
```

```python
y.head()
```

```python
mi = mutual_info_regression(X, y)
```

```python
mi
```

```python
# Select the k best features.
selector = SelectKBest(mutual_info_regression, k=5)
selector.fit(X, y)

# Get indices of columns to keep.
idxs_selected = selector.get_support(indices=True)

# Create new DataFrame with only desired columns.
X_new = X.iloc[:, idxs_selected]

# Show the top 5 features.
print(X_new.head())
```

Sale price should increase as values of top-ranked features go up. Based on our top-5 features, older houses tend to sell at lower price. Overall quality of house also has effects on sale price, better quality implies higher sale price. Housing size also contributes to sale price as Total Bsmt SF, Gr Liv Area, and Garage Area become bigger, sale price will increase. Overall, we can see there's a positive correlation bewteen sale price and top-5 ranked features.
