# A function that given a user id, returns: (1) this user's age and occupation, (2) the average rating given by this user, and (3) his highest and lowest rated movie(s)

```python
data1 = pd.read_csv('movie_lens_genre_separated.tsv', sep='\t')
```

```python
data1.head()
```

```python
def user_info(user_id):
    specific_user = data1.loc[data1['user_id'] == user_id]
    if specific_user.empty:
        return 'Could not find a record with given user ID'
    specific_user_age =specific_user["age"].values[0]
    specific_user_occupation_index =specific_user["occupation"].values[0]
    specific_user_occupation = occupation_dict[specific_user_occupation_index]
    specific_user_rating = specific_user["rating"].mean(skipna = True)
    highest_rate_num = specific_user["rating"].max()
    highest_rate = specific_user.loc[specific_user['rating'] == highest_rate_num]
    highest_movie_series = highest_rate.drop_duplicates(subset=['title'])
    highest_movie_list = highest_movie_series['title'].tolist()
    lowest_rate_num = specific_user["rating"].min()
    lowest_rate = specific_user.loc[specific_user['rating'] == lowest_rate_num]
    lowest_movie_series = lowest_rate.drop_duplicates(subset=['title'])
    lowest_movie_list = lowest_movie_series['title'].tolist()
    print("Age:",specific_user_age, ", Occupation:", specific_user_occupation, 
      ", Avg_Movie_Rating:", specific_user_rating, "\nHighest rated movie(s):", highest_movie_list,
     "\nLowest rated movie(s):", lowest_movie_list)
```

```python
user_info(15)
```

```python
user_info(25264626)
```

```python
most_50 = data.groupby('movie_id').size().sort_values(ascending=False)[:50]
```

```python
pivoted = data.pivot_table(index=['movie_id', 'title'],
                           columns=['gender'],
                           values='rating',
                           fill_value=0)
```

```python
pivoted.head()
```

```python
pivoted['diff'] = pivoted.M - pivoted.F
pivoted.head()
```

```python
pivoted.reset_index('movie_id', inplace=True)
```

```python
disagreements = pivoted[pivoted.movie_id.isin(most_50.index)]['diff']
disagreements.sort_values().plot(kind='barh', figsize=[9, 15])
plt.title('Male vs. Female Avg. Ratings\n(Difference > 0 = Favored by Men)')
plt.ylabel('Title')
plt.xlabel('Average Rating Difference');
```

Male likes Godfather more than female, and female likes Rocky Horror Picture Show more than male
